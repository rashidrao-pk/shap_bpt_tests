# Create a copy of the test/ground_truth hazelnut dataset 
with four rotation variations of each sample

```python
import os
from PIL import Image

# Create the test2 and ground_truth2 folders
os.makedirs('test2', exist_ok=True)
os.makedirs('ground_truth2', exist_ok=True)

for classname in ['crack', 'cut', 'hole', 'print']:
    os.makedirs(f'test2/{classname}', exist_ok=True)
    os.makedirs(f'ground_truth2/{classname}', exist_ok=True)

# Run for each file in the classes in test
for classname in ['crack', 'cut', 'hole', 'print']:
    number = 0
    # Enumerate all files in test/{classname}
    for filename in os.listdir(f'test/{classname}'):
        if filename.lower().endswith(('.png')):
            name, ext = os.path.splitext(filename)
            img_path = f'test/{classname}/{name}.png'
            gtruth_path = f'ground_truth/{classname}/{name}_mask.png'

            img = Image.open(img_path)
            gtruth = Image.open(gtruth_path)

            # resize
            img = img.resize((256, 256), resample=Image.BILINEAR)
            gtruth = gtruth.resize((256, 256), resample=Image.NEAREST)

            for rotation in [0, 90, 180, 270]:
                img2 = img.rotate(rotation)
                gtruth2 = gtruth.rotate(rotation)

                # Save the images
                img2.save(f'test2/{classname}/{number:03}.png')
                gtruth2.save(f'ground_truth2/{classname}/{number:03}_mask.png')
                number += 1
```
